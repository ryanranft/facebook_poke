```python
import pandas as pd
import os
from pathlib import Path
import hvplot.pandas
%matplotlib inline
```

```python
nft_sales = pd.read_csv(Path("Resources/nft_sales.csv"))
nft_sales.head()
```

```python
# Grab a slice of top 30 NFTs.  Clean the data removing the "$" and commas from the data and convert to a float type.

nft_sales_30 = nft_sales[0:31]
nft_sales_30['Sales'] = nft_sales_30.Sales.str.replace('$', '')
nft_sales_30['Sales'] = nft_sales_30.Sales.str.replace(',', '').astype(float)
nft_sales_30['Buyers'] = nft_sales_30.Buyers.str.replace(',', '').astype(float)
nft_sales_30['Owners'] = nft_sales_30.Owners.str.replace(',', '').astype(float)
nft_sales_30['Txns'] = nft_sales_30.Txns.str.replace(',', '').astype(float)
nft_sales_30.info()
```

```python
# Preview the data

nft_sales_30.head(20)
```

```python
# Show total Sales of top 30 NFT by Sale Totals.  This data reflects, the power of Utility within the NFT market. and the most sucessful NFT art projects

nft_sales_30a = nft_sales_30.hvplot.bar(
    x='Collections', 
    y= 'Sales',
   # ylim=(0,5000044650),
    rot=90,
    xlabel='NFT Collection as of October 2022', 
    ylabel='Total Sales'
).opts(yformatter='%.0f')
nft_sales_30a
```

```python
# Show total tramsactions of top 30 NFT by volume.  demostrates the level of Risk in investing 

nft_sales_30c = nft_sales_30.hvplot.bar(
    x='Collections', 
    y= 'Txns',
    rot=90,
   # width=500,
   # height=100,
    xlabel='NFT Collection as of October 2022', 
    ylabel='Transactions'
).opts(yformatter='%.0f')
nft_sales_30c
```

```python
nft_history = pd.read_csv(Path("Resources/NFT_Sales_History.csv",index_col="Date", parse_dates=True, infer_datetime_format=True))
nft_history.head()
```

```python
nft_history.info()
```

```python
# NFT sale history data from 2017 to 2021.  Dropping unneeded columns. setting the index and preparing for a time-series analysis.  
# Originally the data appeared to not have much information for the first 1400 rows of information.  
# In actuality the full data set proves and reflects the volatility and the hype cycle of NFTs.


nft_history_slim = nft_history.drop(columns=['Sales_USD_cumsum',
                                             'Number_of_Sales_cumsum',
                                             'Active_Market_Wallets_cumsum',
                                             'Primary_Sales_cumsum',
                                             'Secondary_Sales_cumsum',
                                             'AverageUSD_cum',
                                             'Active_Market_Wallets',
                                             'Primary_Sales'
                                        ])
nft_history_timeframe = nft_history_slim # [1400:]
nft_history_timeframe.set_index(pd.to_datetime(nft_history_timeframe['Date'], infer_datetime_format=True), inplace=True)
nft_history_timeframe.info()
```

```python
# In order to make the data more clear, I needed to change the timeframe to monthly intervals by totalling the days in the month.

nft_history_timeframe.head()
nft_history_sum = nft_history_timeframe.groupby(nft_history_timeframe.index.strftime('%Y-%m')).sum()
nft_history_sum
```

```python
#  lineplot data using a width of 1200 to add clearity. 

nft_history_sum.hvplot.line(
    x='Date', 
    y='Sales_USD',
    rot=90,
    width=1200,
    xlabel='June 2017 to November 2021', 
    ylabel='Sales in USD'
).opts(yformatter='%.0f')
```
